```python
# -*- coding: utf-8 -*-

# -*- coding: utf-8 -*-
"""
Spyder Editor

This is a temporary script file.
"""

import pandas as pd
import numpy as np


#Read data from SAP file 
#file_sfdc = "C:\Users\Bedantika\Documents\Python Training\test_sfdc"
#dfsap = pd.read_csv(r"C:\Users\Bedantika\Documents\Python Training\test_sap.csv")
#dfsap = pd.read_csv(r"C:\Users\Bedantika\Recon\sap_output.csv")
dfsap = pd.read_csv(r"C:\Users\Bedantika\Recon\sap_output2019_07_22.csv")

#Read data from SFDC file 
print(dfsap)
#dfsfdc = pd.read_csv(r"C:\Users\Bedantika\Documents\Python Training\test_sfdc.csv")
#dfsfdc = pd.read_csv(r"C:\Users\Bedantika\Recon\sfdc_output.csv")
dfsfdc = pd.read_csv(r"C:\Users\Bedantika\PycharmProjects\SOQL_datadump\datafile\Quote.csv")

print(dfsfdc)
print("****")

#Joining data from 2 files with 'Name' as key 
#common = pd.merge(dfsap,dfsfdc,on = 'Name',how = 'outer',indicator = True)

#common = pd.merge(dfsap,dfsfdc,on = 'ZZSFDC_QUOTE_ID',how = 'outer',indicator = True)
common = pd.merge(dfsap,dfsfdc,on = 'QUOTE_ID',how = 'outer',indicator = True)
print(common)

print("****")
print("****")

# Listing the columns of a file 
l = list(dfsap.columns)

#Removing the primary key from the list to be comapred between files 
#l.remove('Name')   
l.remove('QUOTE_ID')
#l.remove('ZZSFDC_QUOTE_ID')

print(l)

def check(row):

     mismatch_flg = 0 

# Records are present only in the SAP file 
     if  row['_merge'] == 'left_only' :
          return 'SAP_Only'
        
# Records are present only in the SFDC file         
     elif row['_merge'] == 'right_only' :
         return 'SFDC_Only'
                
# Records are present in both files and hence being compared fieldwise
     elif row['_merge'] == 'both' :
        
            s = "Data match:"
            
# Create the column names dynamically and compare their values             
            for column in l:

                colx = column + '_x'
                coly = column + '_y'     

                if row[colx] != row[coly] :
                    
                    if mismatch_flg == 0 :
                        mismatch_flg = 1 
                        s = "Data Mismatch"

                    s = s  + column + ','

            if s != "":
                return s 

        
# Apply 'check' fundtion for each record and put compare result in comment field
# comment field will have SAP_only for records present only in SAP file 
# SFDC_only for records only in SFDC file and field indicators for mismatch for records in both files

for  row in common.itertuples(): 
        common['comment'] = common.apply(lambda row : check(row), axis = 1) 

# Removing the '_merge' column from display 
common.drop('_merge', axis=1, inplace=True)


#Saving the merged data to a file 
common.to_csv('recon_output.csv')

print(common)
```
